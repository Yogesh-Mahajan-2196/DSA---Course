```python
class Solution:
    def solve(self, head, n):
        p1 = head
        p2 = head

        for i in range(n):
            p2 = p2.next

        if p2 == None:
            head = head.next
            return head

        while p2.next != None:
            p1 = p1.next
            p2 = p2.next

        p1.next = p1.next.next

        return head.val
    
sol = Solution()
print(sol.solve(head,3))
```

Remove Duplicate for S0rted List - 83

```python
class Solution:
    def solve(self, head):
        curr = head
        if curr == None or curr.next == None:
            return curr
        
        while curr.next != None:
            if curr.next.val == curr.val:
                curr.next = curr.next.next
            else:
                curr = curr.next

        return head

sol = Solution()
print(sol.solve(head))
```
