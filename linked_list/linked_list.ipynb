```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
a = Node(5)
b = Node(3)
c = Node(7)
```

```python
head = a
print(head.data)
```

```python
a.next = b
b.next = c
```

```python
print(head.next.data)
```

```python
print(head.next.next.data)
```

```python

def printLinkedList(head):
    curr = head
    while curr != None:
        print(curr.data, end=' ')
        curr = curr.next
printLinkedList(head)
```

Insert At First Node

```python
new_node = Node(4)
new_node.next = head
head = new_node
```

```python
printLinkedList(head)
```

```python
new_node = Node(1)
curr = head
print(curr.data)
```

Insert at Last Node

```python
while curr.next != None:
    curr = curr.next
curr.next = new_node
```

```python
printLinkedList(head)
```

Insert at Kth Index

```python
new_node =Node(6)
k = 3
curr = head
for i in range(k-1):
    curr = curr.next

new_node.next = curr.next
curr.next = new_node
```

```python
printLinkedList(head)
```

Delete First Node

```python
head = head.next
printLinkedList(head)
```

Delete Lat Node

```python
curr = head
while curr.next.next != None:
    curr = curr.next

curr.next = None
printLinkedList(head)
```

Delete Kth Node

```python
k = 1
curr = head
for i in range(k-1):
    curr = curr.next
curr.next = curr.next.next

printLinkedList(head)
```
