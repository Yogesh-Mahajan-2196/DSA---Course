```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

a = Node(2)
b = Node(3)
c = Node(5)
```

```python
head = a
a.next = b
b.next = c
b.prev = a
c.prev = b
```

```python
def print_doubly(head):
    curr = head
    while curr != None:
        print(curr.data, end= " ")
        curr = curr.next

print_doubly(head)
```
