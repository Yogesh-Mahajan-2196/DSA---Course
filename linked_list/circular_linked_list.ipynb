```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

a = Node(3)
b = Node(4)
c = Node(5)
d = Node(6)
```

```python
head = a
a.next = b
b.next = c
c.next = d
d.next = a
a.prev = d
b.prev = a
c.prev = b
d.prev = c
```

```python
def print_circular_linked_list(head):
    curr = head
    while True:
        print(curr.data, end=' ')
        curr = curr.next
        if curr == head:
            break

print_circular_linked_list(head)
```
